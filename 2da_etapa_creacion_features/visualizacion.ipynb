```python
import pandas as pd

df = pd.read_csv('seismic_features.csv')


len(df)
```

```python
# Visualizar primeros 20 mainshocks (es_mainshock = 1)
if 'es_mainshock' not in df.columns:
    raise ValueError("La columna 'es_mainshock' no está presente en el DataFrame. Asegúrate de haber generado el nuevo seismic_features.csv con esa columna.")

mainshocks = (
    df[df['es_mainshock'] == 1]
      .sort_values('Date(UTC)')
      .head(20)
      [['id_sismo_principal','Date(UTC)','Latitude','Longitude','Depth','Magnitude','es_mainshock','ventana_temporal_replica','existe_replica_fuerte']]
)

print(f"Total mainshocks (M>=5.5 y criterio de precedencia): {df['es_mainshock'].sum()}")
mainshocks
```
